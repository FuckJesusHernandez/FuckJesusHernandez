```python
comidas = ["Pizza", "Hamburguesa", "Sándwich", "Ensalada", "Pasta"]
bebidas = ["Café", "Té", "Refresco", "Agua", "Jugo"]
alumnos = []

while True:

    print("\nMenú de opciones:")
    print("1. Agregar alumno")
    print("2. Agregar pedido")
    print("3. Imprimir todos")
    print("4. Historial de comida")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        nombre_alumno = input("Ingrese el nombre del alumno: ")
        alumnos.append(nombre_alumno)
        print(f"{nombre_alumno} ha sido agregado a la lista de alumnos.")

    elif opcion == "2":
        if len(alumnos) == 0:
            print("Primero debe agregar un alumno.")
        else:
            print("Tipos de comidas disponibles:")
            for i in range(len(comidas)):
                print(f"{i + 1}. {comidas[i]}")

            comida_index = int(input("Seleccione el número de la comida: ")) - 1

            print("Tipos de bebidas disponibles:")
            for i in range(len(bebidas)):
                print(f"{i + 1}. {bebidas[i]}")

            bebida_index = int(input("Seleccione el número de la bebida: ")) - 1

            print("Lista de alumnos:")
            for i in range(len(alumnos)):
                print(f"{i + 1}. {alumnos[i]}")

            alumno_index = int(input("Seleccione el número del alumno: ")) - 1

            pedido = {
                "Comida": comidas[comida_index],
                "Bebida": bebidas[bebida_index],
                "Alumno": alumnos[alumno_index]
            }

            alumnos.pop(alumno_index)
            alumnos.insert(alumno_index, pedido)
            print("Pedido agregado con éxito.")

    elif opcion == "3":

        print("Lista de comidas disponibles:")
        for comida in comidas:
            print(comida)

        print("Lista de bebidas disponibles:")
        for bebida in bebidas:
            print(bebida)

        print("Lista de alumnos y sus pedidos:")
        for alumno in alumnos:
            if isinstance(alumno, dict):
                print(f"Alumno: {alumno['Alumno']}, Comida: {alumno['Comida']}, Bebida: {alumno['Bebida']}")
            else:
                print(f"Alumno: {alumno}")

    elif opcion == "4":

        print("Historial de comidas:")
        for comida in comidas:
            print(comida)

    elif opcion == "5":

        break

    else:
        print("Opción no válida. Por favor, seleccione una opción válida.")
```
